```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Housing.csv")
data.head()
```

```python
data.info()
```

```python
# Line Plot

plt.plot(data["price"],data["area"])
```

```python
plt.plot(data["bedrooms"],data["price"])
```

```python
plt.plot(data["furnishingstatus"],data["price"])
```

```python
plt.bar(data["area"],data["price"])
plt.xlabel("Area")
plt.ylabel("Price")
```

```python
plt.bar(data["bedrooms"],data["price"],label="Price")
plt.legend()
plt.ylabel("Price")
plt.xlabel("Bedrooms")
```

```python
plt.scatter(data["bedrooms"],data["stories"])
plt.ylabel("Stories")
plt.xlabel("Bedrooms")
```

```python
sns.barplot(y="bedrooms",data=data,hue="bathrooms")
```

```python
sns.countplot(x="bedrooms",data=data,hue="parking",palette="deep")
```

```python
sns.boxplot(x="price",data=data)
```

```python
sns.histplot(x="bedrooms",data=data,hue="parking",palette="deep")
```

```python
d=data[["parking"]]
sns.heatmap(d,cmap="icefire",linecolor="g",linewidths=2)
```
