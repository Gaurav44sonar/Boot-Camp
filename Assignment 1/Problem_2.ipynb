```python
import pandas as pd
```

```python
data=pd.read_csv("Sales.csv")
data.head()
```

```python
data["New_Revenue"]=data["Price"]*data["QuantitySold"]
data
```

```python
data.describe()
```

```python
z_score=abs((data["Price"]-data["Price"].mean())/(data["Price"]).std())
z_score
```

```python
data["z_score"]=z_score
data
```

```python
data=data[data["z_score"]<2]
data
```

```python
data=data.drop("z_score",axis=1)
data
```
